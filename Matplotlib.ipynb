```python
import matplotlib.pyplot as plt

x = ['Mon','Tues','Wed','Thur','Fri']
y = [10,15,7,18,5]

plt.plot(x,y)

plt.title('Bakery of This week')
plt.xlabel('Day of the week')
plt.ylabel('Sales per day')

plt.show()
```

```python
#plt.plot(x,y,color - 'color name' , linnestyle = 'line_stayle' , linewit = '')

import matplotlib.pyplot as plt
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales = [100,200,2000,400,3000,600,456,345,2345,1233,2344,345]

plt.plot(months,sales,color = 'blue', linewidth = 2, marker = 'o', label = 'Sales', linestyle = '--')
plt.xlabel('Months')
plt.ylabel('Sales Per Month')
plt.title('Monthly By Sales Report')
plt.legend()
plt.grid(color = 'yellow',linewidth = 1,linestyle = '--',)


plt.show()
```

```python
# plt.bar(x,height,color name, width=value, label='label name')
import matplotlib.pyplot as plt

product = ['A','B','C','D','E']
sales = [1000,200,2000,400,1500]

plt.bar(product,sales,color='orange',label='sales 2025')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('product sale comparoson')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

regions = ['North','South','East','West']
sales = [1000,1600,2000,1500]

plt.pie(sales,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','coral'])
plt.title('Sales Report')
plt.show()
```
