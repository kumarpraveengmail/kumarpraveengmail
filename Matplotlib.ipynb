```python
import matplotlib.pyplot as plt
import pandas as pd

#read csv file
df = pd.read_csv('/content/netflix_titles.csv')
df.head()
```

```python
#Clean data
df = df.dropna(subset=['type','release_year','rating','country','duration'])

type_counts = df['type'].value_counts()
plt.bar(type_counts.index,type_counts.values,color=['orange','red'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Distribution of Movie and TV Shows')
plt.show()

rating_counts = df['rating'].value_counts()
#plt.figure(rating_counts.index,rating_counts.index, autopct='%1.1f%%',startangle=90)
plt.pie(rating_counts.values,labels=rating_counts.index,autopct='%1.1f%%',startangle=90)

plt.title('Distribution of Ratings')


movie_df = df[df['type'] == 'Movie']
movie_df['duration_int'] = movie_df['duration'].str.replace('min','').astype(int)

plt.figure(figsize=(10,6))
plt.hist(movie_df['duration_int'],bins=20,color='green',edgecolor='black')
plt.xlabel('Duration (min)')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Durations')
plt.show()

release_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(release_counts.index,release_counts.values,color='blue',marker='o')
plt.xlabel('Release Year')
plt.ylabel('Number of count')
plt.title('Release Year Distribution')
plt.show()

country_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(10,6))
plt.barh(country_counts.index,country_counts.values,color='teal')
plt.xlabel('Number of count')
plt.ylabel('Country')
plt.title('Top 10 Countries with Most Content')
plt.show()

content_by_year = df.groupby(['release_year','type']).size().unstack().fillna(0)

fig,ax = plt.subplots(1,2, figsize=(10,6))

#first subplot : movies
ax[0].plot(content_by_year.index,content_by_year['TV Show'],marker='o',linestyle='-',color='blue')
ax[0].set_xlabel('Release Year')
ax[0].set_ylabel('Number of Movies')
ax[0].set_title('Number of Movies by Release Year')

#first subplot : TV show
ax[0].plot(content_by_year.index,content_by_year['Movie'],marker='*',linestyle=':',color='orange')
ax[0].set_xlabel('Release Year')
ax[0].set_ylabel('Number of TV Show')
ax[0].set_title('Number of TV Show by Release Year')
legend = ax[0].legend(['TV Show','Movie'])

plt.tight_layout()
plt.show()
```
